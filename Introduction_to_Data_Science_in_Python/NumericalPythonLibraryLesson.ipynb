# Array Operations

```python

a = np.array([10,20,30,40])
b = np.array([1,2,3,4])

c = a-b
print(c)

d = a*b
print(d)
```

```python
#Temperature excercise

farenheit = np.array([0,-10,-5,-15,0])

#celcius = (F -32) * 5/9

celsius = (farenheit -32) * (5/9)
celsius
```

```python
#To check temps greater than -20

celsius > -20
```

```python
#To check the even temps
celsius%2 == 0
```

```python
#Elementwise manipulation 

#Elementwise product
A = np.array([[1,1], [0,1]])
B = np.array([[2,0], [3,4]])
print(A*B)

#Matrix product uses @ or use the dot function
print(A@B)
```

```python
#The product of two matrices is only plausible when the inner dimensions of both are the same. 

A.shape
```

```python
#When manipulating two arrays, numpy generalizes the type. This is called upcasting.

array1 = np.array([[1,2,3], [4,5,6]])
print(array1.dtype)

array2 = np.array([[7.1,8.2,9.3], [10.4,11.5,12.6]])
print(array2.dtype)
```

```python
array3 = array1 + array2
print(array3)
print(array3.dtype)
```

```python
#Array aggregation functions
print(array3.sum())
print(array3.max())
print(array3.min())
print(array3.mean())
```

```python
#Para arrays 2D se puede hacer lo mismo para cada fila o columna

b = np.arange(1,16,1).reshape(3,5) #A una forma de 3x5
print(b)
```
